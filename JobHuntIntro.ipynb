<p>As an undergraduate, I was exposed to data science in a course called "Data Mining and Visualization". Then again, as a graduate student, data science was brought up in a course called "Big Data Analytics and Management". There is a famous data science Venn diagram and many variations of it. The link to an article and the diagram can be found here: <a href='http://drewconway.com/zia/2013/3/26/the-data-science-venn-diagram'>Data Science Venn Diagram</a>. Because I am doing this in a Jupyter Notebook, I will recreate this diagram with Python.</p>

```python
#import module 
from matplotlib_venn import venn3, venn3_circles 
from matplotlib import pyplot as plt 
  
# depict venn diagram 
v = venn3(subsets=(20, 20, 5, 20, 5, 5, 5),  
          set_labels=('Computer Science', 'Statistics and Math', 'Business/Domain Knowledge')) 

v.get_label_by_id("100").set_text("C.S.") 
v.get_label_by_id("101").set_text("Apps") 
v.get_label_by_id("111").set_text("D.S.") 
v.get_label_by_id("001").set_text("Business Knowledge")
v.get_label_by_id("011").set_text("Analysis")
v.get_label_by_id("010").set_text("Stats/Math")
v.get_label_by_id("110").set_text("M.L.")

# add outline 
venn3_circles(subsets=(20, 20, 5, 20, 5, 5, 5),  
              linestyle="solid", linewidth=2) 
  
# assign title 
plt.title("A Data Science Venn Diagram") 
plt.show() 
```
